```python
import numpy as np
import matplotlib.pyplot as plt

result = {
  "resnet50(baseline)": [39.21, 0.679, "o"],
  "resnet18": [37.32, 0.64, "x"],
  "resnet50-Poolformer": [33.77, 0.60, "*"],
}
for name, res in result.items():
  plt.plot(res[0], res[1], marker=res[2], label=name)

plt.title(f'BEVFusion')
plt.xlabel("Time(ms)")
plt.ylabel('mAP')
plt.legend(loc="upper left")
plt.show()
```

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

matplotlib.rc("font", family="Times New Roman")
plt.rcParams["font.family"] = "Times New Roman"
# plt.rcParams["font.size"]
# plt.rcParams["font.style"]

result = {
    # Time(ms), mAP
    "PV-RCNN": [4, 67.0, "o", "gold"],
    "PV-RCNN++": [9, 69.5, "o", "purple"],
    "CenterPoint": [25, 62.0, "o", "green"],
    "CenterFormer": [5, 61.0, "o", "lime"],
    "DSVT": [10, 72.0, "o", "crimson"],
    "PillarNext": [30, 62.0, "o", "dodgerblue"],
    "ECF(ours)": [31, 61.7, "o", "navy"]
}
for name, res in result.items():
    plt.plot(res[0], res[1], marker=res[2], color=res[3], label=name)
    plt.annotate(name, (res[0], res[1]), ha="center", va="top")

font = {
    "family": "Times New Roman", "fontweight": "bold"
}
plt.xlabel("Scene per second (Hz)", fontdict=font)
plt.ylabel("mAP", fontdict=font)
# plt.legend(loc="upper left")
plt.show()
```
